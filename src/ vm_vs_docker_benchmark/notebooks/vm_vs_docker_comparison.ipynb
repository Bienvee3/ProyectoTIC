```python
import pandas as pd
import matplotlib.pyplot as plt

vm = pd.read_csv("../results/benchmark_snake_vm.csv")
docker = pd.read_csv("../results/benchmark_snake_docker.csv")

plt.figure(figsize=(10,5))
plt.plot(vm["cpu_percent"], label="VM CPU")
plt.plot(docker["cpu_percent"], label="Docker CPU")
plt.legend()
plt.title("Uso de CPU: VM vs Docker")
plt.show()
```
